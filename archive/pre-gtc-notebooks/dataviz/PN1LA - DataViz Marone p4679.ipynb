```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style="ticks")
```

```python
GTC_DATA_DIR = "/gws/nopw/j04/ai4er/users/pn341/earthquake-predictability/data/gtc_quakes_data"
LABQUAKES_DATA_DIR = f"{GTC_DATA_DIR}/labquakes"
MARONE_DATA_DIR = f"{LABQUAKES_DATA_DIR}/Marone"
```

```python
p4679_FILE_PATH = f"{MARONE_DATA_DIR}/p4679/p4679.txt"
with open(p4679_FILE_PATH, "r") as file:
    lines = [next(file) for _ in range(10)]
print("".join(lines))
```

```python
with open(p4679_FILE_PATH, "r") as file:
    df = pd.read_csv(file, skiprows=1)

df.columns = [
    "id",
    "lp_disp",
    "shr_stress",
    "nor_disp",
    "nor_stress",
    "time",
    "mu",
    "layer_thick",
    "ec_disp",
]

df = df.drop(["id"], axis=1)
```

```python
df.head()
```

```python
subset_df = df[(df["time"] >= 4450) & (df["time"] <= 4460)].copy()
subset_df.shape
```

```python
plt.figure()
sns.lineplot(data=subset_df, x="time", y="lp_disp")
plt.xlabel("time [s]")
plt.ylabel("lp_disp [mic]")
plt.show();
```
